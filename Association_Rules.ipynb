```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
retail_data = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Files/Online retail.xlsx',header=None,)
```

### Preprocessing

```python
retail_data = retail_data.rename(columns={0:'Products'})
```

```python
retail_data_3 =  retail_data['Products'].str.split(',').tolist()
retail_data_3[:5]
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
encoder = TransactionEncoder().fit(retail_data_3)
onehot = encoder.transform(retail_data_3)
onehot = pd.DataFrame(onehot,columns=encoder.columns_)
onehot.head()
```

#### Apriori Algorithm

```python
frequent_itemsets_2 = apriori(onehot,min_support=0.009,use_colnames=True)
frequent_itemsets_2
```

```python
rules2 = association_rules(frequent_itemsets_2,metric='lift',min_threshold=1)
rules2
```

```python
# TOP 20 ASSOCIATION RULES
rules2.sort_values(by='lift',ascending=False).head(20)
```

> What is Lift and Why is it important in Association Rules?
* Lift is the Measure of how strong is the Association betweeen antecedant and consequent.
* It is a measure which takes into the account bias that is introduced in the association purly by chance.

> What is Support and Confidence? How do we calculate them?
* Support is the proportion of transactions in the dataset that contain both the antecedent and the consequent..
* Support =
Number of transaction with Antecedent and consequent / Total Transactions
* Confidence measures the likelihood that the consequent will be purchased when the antecedent is purchased.
* Confidence = Number of transaction with Antecedent and consequent / total andecedent
* Can also be given by :
  *Support = AnB/N
  *Confidence = AnB/A

> What are some limitations or challanges of association rules mining?
* Association rules mining are not best when the dataset is large with multiple features
* Association mining can be carried out only for categorical data.
* Determining the threshold value for minimum support value in apriori algorithm is only by trial and error.
* Sometimes the association that are found can be trivial or be inexplicable. Idnetifying the actionable Rules is important out of all the rules
